# Bananes et filtre orange
## filtre orange
Le programme suivant traite chaque pixel de position (x,y), ou plutôt les composantes RVB de chaque pixel avec la fonction de transfert du filtre orange. 
Pour chaque pixel, la donnée numérique de la composante rouge est stockée dans `p[0]`, celle verte dans `p[1]`, et celle bleue dans `p[2]`.
Le filtre orange sera : R: 80%, V:50%, B: 0%
Multiplier par 80% la composante rouge du pixel revient alors à faire : `p[0]*0.8`.
## à faire
* Executer chacune des cellules suivantes pour transformer l'image des bananes en une image filtrée en orange.
* On pourra modifier les paramètres du filtre (les coefficients 0.8; 0.5; 0) pour obtenir un effet encore plus orange.



```python
from PIL import Image
```

```python
imageSource=Image.open("bananes.jpg")
largeur,hauteur=imageSource.size
planPixels=Image.new("RGB",(largeur,hauteur))

for y in range(hauteur): # y varie de 0 à hauteur - 1
    for x in range(largeur): # x varie de 0 à largeur - 1
        p=imageSource.getpixel((x,y)) # p est la valeur RGB du pixel
        p = (int(p[0]*0.8), int(p[1]*0.5), p[2]*0)
        planPixels.putpixel((x,y),p) # nouvelle abscisse de 0 à h-1

planPixels.save("banane_filtre.jpg")
planPixels.show()
```

# traitement d'une photographie astronomique
##  quel est le problème avec la photographie astronomique ?
Les photographies astronomiques sont prises avec des réglages particuliers. A cause du manque de lumière, les temps de pose sont plus longs. Et les capteurs numériques sont très sensibles aux radiations rouge émises par l'hydrogène. Les photographies sont alors trop "rouges".



## à faire
Modifier le programme suivant pour appliquer un filtre qui aura pour effet de :
* diminuer la composante rouge de tous les pixels de l'image
* d'en augmenter la composante bleue

Attention, prendre des valeurs à peine supérieures et inférieures à 1.0 pour ces traitements, qui doivent rester assez legers...


```python
imageSource=Image.open("ciel profond.jpg")
largeur,hauteur=imageSource.size
planPixels=Image.new("RGB",(largeur,hauteur))

for y in range(hauteur): # y varie de 0 à hauteur - 1
    for x in range(largeur): # x varie de 0 à largeur - 1
        p=imageSource.getpixel((x,y)) # p est la valeur RGB du pixel
        p = (int(p[0]*1.0), p[1], int(p[2]*1.0))
        planPixels.putpixel((x,y),p) # nouvelle abscisse de 0 à h-1

planPixels.save("ciel profond corrige.jpg")
planPixels.show()
```
